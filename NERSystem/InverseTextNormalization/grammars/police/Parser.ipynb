```python
import re
from sys import argv, exit
#sentence = "<EMERGENCY>  harassing  </EMERGENCY>  <O> occured at </O> <Addr-Building-Number> one eight seven </Addr-Building-Number>  <Addr-Building-Name> waterloo hostel</Addr-Building-Name> <Addr-Street> ellington square </Addr-Street> <Addr-UnitNo> hash sixteen dash three hundred forty </Addr-UnitNo> <PostCode> eight four five two nine nine  </PostCode> <Addr-city> Singapore </Addr-city>" 
#sentence="<O> reported at </O> <EMERGENCY> harassing </EMERGENCY> <Addr-Block> block four </Addr-Block> <Addr-Street> mimosa road </Addr-Street> <Addr-UnitNo> hash five dash three hundred nine </Addr-UnitNo> <PostCode> six four five two zero five  </PostCode> <Addr-city> Singapore </Addr-city>"
def get_tagpairs(sentence):
    sentence = sentence.strip()
    wordlist = re.split(r'(?<=>)(.+?)(?=<)', sentence)
    print(wordlist)
    raw_taglist = wordlist[::4]

    taglist = [x[1:-1] for x in raw_taglist]

    phrases = map(str.strip, wordlist[1::4])
    tagged = map(lambda x: tag_items(x[0], x[1]), zip(taglist, phrases))

    return tagged


def tag_items(tag, phrase):
    HEADER = 'B-{}'.format(tag)
    TRAILER = 'L-{}'.format(tag)
    MIDDLE = 'I-{}'.format(tag)
    SINGLE = 'U-{}'.format(tag)

    words = [x.lower() for x in phrase.split()]

    if tag == 'O':
        return [[x, 'O'] for x in words]
    
    if len(words) == 0:
        return [[x, 'O'] for x in words]
    
    if len(words) == 1:
        return [[words[0], SINGLE], ]

    if len(words) == 2:
        return [[words[0], HEADER], [words[1], TRAILER]]

    intermediaries = [[x, MIDDLE] for x in words[1:-1]]

    result = []
    result.extend([[words[0], HEADER], ])
    result.extend(intermediaries)
    result.extend([[words[-1], TRAILER], ])

    return result

```

```python
file = "testrunclean.txt"
outfile = "trainrun.bilou"


#file = "police.txt"
#outfile = "trainpolice.bilou"
#outfile2= "testpolice.txt"
```

```python
with open(file) as f:
        data = f.read()
print(data)
```

```python
ldata = [x for x in data.split('\n') if x]
#ldata = [sentence]
```

```python
final = []
for x in ldata:
    x=x.lstrip(" ")
    if x[0] != "<":
        continue
#    print(sum(get_tagpairs(x),[]))
    raw_processed = sum(get_tagpairs(x), [])
    processed = [' '.join(x) for x in raw_processed]
    final.append(processed)
```

```python
print(final)
```

```python
DELIMITER = '\n\n-DOCSTART- O\n\n'
final = ['\n'.join(x) for x in final]

final_output = DELIMITER[2:] + DELIMITER.join(final) + '\n'
print(final_output)
with open(outfile, 'w') as f:
    f.write(final_output)
```
